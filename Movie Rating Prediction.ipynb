```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IMDb Movies India.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(method = 'ffill', inplace = True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

**Removing the paranthesis from the dataset**

```python
df['Year'] = df['Year'].str.replace(r'[()]', '', regex = True).astype(int)
```

**Removing the word 'min' from the attribute 'Duration'**

```python
df['Duration'] = pd.to_numeric(df['Duration'].str.replace(' min', ''))
```

```python
df.head()
```

**Keeping the unique values of genres and replacing the null values with mode**

```python
df['Genre'] = df['Genre'].str.split(', ')
```

```python
df = df.explode('Genre')
df['Genre'].fillna(df['Genre'].mode()[0], inplace = True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

**Dropm Column 'Name' as Linear Regression should not have any 'Object' data types**

```python
df.drop('Name', axis = 1, inplace = True)
```

```python
df.head()
```

```python
director_rating = df.groupby('Director')['Rating'].transform('mean')
df['Director'] = director_rating
```

```python
df.head()
```

```python
a1_rating = df.groupby('Actor 1')['Rating'].transform('mean')
df['Actor 1'] = a1_rating
```

```python
genre_rating = df.groupby('Genre')['Rating'].transform('mean')
df['Genre'] = genre_rating
```

```python
df['Votes'] = df['Votes'].str.replace(',', '')
df['Votes'] = df['Votes'].str.replace(r'\$(\d+\.\d+)M', lambda x: str(float(x.group(1)) * 1_000_000), regex=True)
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
```

```python
a2_rating = df.groupby('Actor 2')['Rating'].transform('mean')
df['Actor 2'] = a2_rating
```

```python
df.head()
```

```python
a3_rating = df.groupby('Actor 3')['Rating'].transform('mean')
df['Actor 3'] = a3_rating
```

```python
df.head()
```

```python
df['Votes']
```

```python
X = df[['Year', 'Duration','Votes', 'Director', 'Actor 1',
       'Actor 2', 'Actor 3']]
Y = df['Rating']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 45)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Evaluate the model
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
